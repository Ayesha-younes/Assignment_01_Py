### ******************* Arithmetic Operators *********************

Q1. Write a Python program to add two numbers and print the result.

```python
# Answer:
# *********** Marks Calculation **********
  
oop_mid = int(input("Enter Mid Marks: "))
oop_terminal = int(input("Enter Terminal Marks: "))

# '+' used for Addition (Arithmetic Operator) 

total_marks = oop_mid + oop_terminal 
print(f"The total of {oop_mid} and {oop_terminal} is = {total_marks}")
```

Q2.Write a program to find the remainder when one number is divided by another.

```python
# Answer:

num_1 = int(input("Enter first number: "))
num_2 = int(input("Enter second number: "))

# '%' used for Remainder (Arithmetic Operator)

remainder = num_1 % num_2
print(f"Remainder when {num_1} is divided by {num_2} is: {remainder}")
```

Q3. Write a program to calculate the square of a number using the power operator.

```python
# Answer:

number = int(input("Enter a number: "))

# '**' Power Operator (Arithmetic Operator)

sq_number = number ** 2
print(f"Square of {number} is: {sq_number}")
```

Q4. Write a program that assigns a value to a variable and then increases it by 5 using an assignment operator.

Q7. Write a program to check whether two numbers are  equal or not.

```python
# Answer:

num_1 = int(input("Enter First number: "))
num_2 = int(input("Enter Second number: "))
print(f"""First number = {num_1}
Second number = {num_2}""")

# '==' Comparison Operator

print(num_1 == num_2)
```

Q8. Write a program to check if one number is greater than another.

```python
# Answer:

num_1 = int(input("Enter First number: "))
num_2 = int(input("Enter Second number: "))
print(f"""First number = {num_1}
Second number = {num_2}""")

# '>' Comparison Operator

print(num_1 > num_2)
```

Q9. Write a program to compare two numbers and print whether the first number is less or equal to the second number.

```python
# Answer:

num_1 = int(input("Enter first number: "))
num_2 = int(input("Enter second number: "))

# '<=-' Comparison Operator

if num_1 <= num_2:
    print(f"{num_1} is less than or equal to {num_2}")
else:
    print(f"{num_1} is greater than {num_2}")    
```

Q10. Write a program that checks if a number is greater than 5 and less than 20.


```python
# Answer:

value_1 = int(input(f"Enter a number: "))

# 'and' Logical operator

if value_1 >5 and value_1 < 20:
    print(f"{value_1} is greater than 5 and less than 20")
else:
    print(f"{value_1} is not in the range")    
```

Q11. Write a program that checks if a number is less than 10 or greater than 50.


```python
# Answer:

value_2 = int(input("Enter a number: "))
print(f"The number is: {value_2}")

# 'or' Logical operator

print(value_2 < 10 or value_2 > 50)
    
```

### ******************* Membership Operators *******************

Q13. Write a program to check if a number exists in a given list.

```python
# Answer:

numbers = [5, 10, 15, 20, 25, 30]
search = int(input("Enter a number to search: "))
print(f"list = {numbers}")
print(f"The number is: {search}.")

# 'in' Membership Operator

print(search in numbers)
```

Q14. Write a program to check if a word exists in a given string.

```python
sentence = "Python Assignment Two"
word = input("Enter a word to search: ")
print(f"Sentence = {sentence}")
print(f"The word is: {word}.")

# 'in' Membership Operator

print(word in sentence)
```

Q15. Write a program to check whether a value is not present in a tuple.


```python
# Answer:

colors_tuple = ("red", "blue", "yellow", "green")
check_color = input("Enter a color to search: ")
print(f"Colors = {colors_tuple}")
print(f"The color is: {check_color}")

# 'not in' Membership Operator

print(check_color not in colors_tuple)
```
