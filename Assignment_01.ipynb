## Variables and Basic Data Types..

### 1.Variable:
   -> A variable in Python is a named storage location that holds a value, which can be changed during program execution.It acts as a container for data.

   -> Syntax to create a variable: variable_name = value

```python
#Example:
student_age = 25
print(f"Age of Student is: {student_age}")
```

### 2.Rules of Variable Naming:
  -> Three rules for naming variables in Python:   
       1.Variable names must start with a letter, or an underscore(_), not a number.  
       2.Variable name can contain letters, numbers, and underscore, but no special characters like @, #, or spaces.  
       3.Variable names are case-sensitive (e.g., "Age" and "age" are different).  

  -> Correct example: user_age = 25  
  -> Incorrect example: 2user = 25 (Start with a number, which is invalid)

### 3.Integer Data Type:
 -> An integer data type(int) represents whole numbers without a decimal point, such as positive, negative, or zero values.

```python
#Example:
student_marks = 90
print(type(student_marks))
print(f"Student Marks = {student_marks}")
```

### 4.Float Data Type:
-> A float data type represents numbers with a decimal point or fractional part, used for real numbers.

```python
#Example:
person_height = 5.7
print(type(person_height))
print(f"Height of a person is: {person_height}")
```

### 5.String Data Type:
-> A string data type (str) represents a sequence of characters, such as text,enclosed in quotes.It can include letters, numbers, symbols, or spaces.

-> Two ways to create a string:  
    1. Using double quotes.  
    2. Using multi-line string (triple quotes).

```python
#Example:

#************* Using Duoble Quotes ***************

student_name = "Ayesha"
print(type(student_name))
print(f"Student name: {student_name}")

#************* Using Triple Quotes ***************

student_info = """Name: Ayesha 
Class: 10th
Section: A"""
print(type(student_info))
print("Student Information:")
print(student_info)
```

### 6.Boolean Data Type:
-> A boolean data type (bool) represents truth values, used for logical operations and conditions.

-> Possible values: True or False.



```python
#Example:
is_student = True
print(type(is_student))
```

### 7.type() Function:
-> The type() function in Python returns the data type of a variable or value, helping to identify what kind of data it holds.

```python
#Example:
person_age = 20
print(type(person_age))
```

### 8.input() Function:
-> The input() function is used in Python to take user input from the keyboard during program execution.The default data type of the input is string(str).

```python
#Example:
user_name = input("Enter your name:")
print(type(user_name))
print(user_name)
```

### 9.Type Casting:
-> Type casting is the process of converting a value from one data type to another, such as changing a string to an integer or float.


```python
#Example:

#**************** String age ********************

age_str = "25"
print(type(age_str))
print(f"String age: {age_str}")

#**************** Integer age ********************

print("--------------------")
age_int = int(age_str)
print(type(age_int))
print(f"Integer age: {age_int}")
```

### 10.Program-Based Question:
-> Python program that takes input from the user and prints its data type.

```python
#************ Taking input from user *****************

user_name = input("Enter your name: ")
user_age = int(input("Enter your age: "))
user_height = float(input("Enter your height in feet: "))
is_student = input("Are you a student? (True/False): ")

#************ Converting string to boolean ************

is_student = bool(is_student)

#******************* Printing values ******************

print("\n******** Student Information ********")
print(f"Name: {user_name}") 
print(f"Age: {user_age}")
print(f"Height: {user_height}")
print(f"Student Status: {is_student}")

#******************* Printing Data Type ***************

print("\n******** Data Types ********")
print("Type of Name:", type(user_name))
print("Type of Age:", type(user_age))
print("Type of Height:", type(user_height))
print("Type of Student Status:", type(is_student))

#******************************************************
```
