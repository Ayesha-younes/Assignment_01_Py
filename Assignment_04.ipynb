### ****************** *While Loop* ******************

*Question:*  
-> If a Python while True loop contains an if, elif, and else block, and all the conditions are true, what will be printed? Explain with an example.

*Answer:*  
-> In Python, if a while True loop contains an if-elif-else block where all conditions are true, only the first true condition will execute in each iteration. The elif and else blocks will be ignored until the loop repeats.

*Explanation:*   
- In Python, the if-elif-else block checks conditions from top to bottom.
- Only the first condition that is true will run, and the rest will be ignored.
- The while True loop runs infinitely until a break is executed.
- Even if all the conditions inside if-elif-else are true, only the first true block executes in each iteration of the loop.

*Example:*  

```python
while True:
    if True:
        print("IF runs")
    elif True:
        print("ELIF runs")
    else:
        print("ELSE runs")
    break
```

> In one iteration, only one block of if-elif-else runs, even if all conditions are true.

***

*Assignment Example:*

```python
string = "aabbcc"
count = 3
while True:
    if string[0] == 'a':      
        string = string[2:]   
    elif string[-1] == 'b':   
        string = string[:2]   
    else:                     
        count += 1
        break

print(string)
print(count)
```

*Step by Step Execution:*
* Start:
    - string = "aabbcc"
    - count = 3
* First iteration:
    - if string[0] == 'a' -> True
    - remove first two characters
    - string = "bbcc"
    - elif and else are skipped
    - Loop repeats
* Second iteration:
    - if string[0] == 'a' -> False
    - elif string[-1] == 'b' -> False
    - else runs -> count = 4
    - loop breaks

> Even if all conditions are true, only the first condition executes each iteration.

***

*Example:*

```python
text = "ccbb aa"
count = 2

while True:
    if text[0] == 'c':
        text = text[2:]
    elif text[-1] == 'a':
        text = text[:-2]
    else:
        count += 1
        break

print(text)
print(count)
```

*Step by Step Execution:*
* Start:
    - text = "ccbb aa"
    - count = 2
* First iteration:
    - if text[0] == 'c' -> True
    - remove first two characters
    - text = "bb aa"
    - elif and else are skipped
    - Loop repeats
* Second iteration:
    - if text[0] == 'c' -> False
    - elif text[-1] == 'a' -> True
    - remove last 2 characters
    - text = "bb "
    - else is skipped
    - Loop repeats
* Third iteration:
    - if text[0] == 'c' -> False
    - elif text[-1] == 'a'-> False
    - else runs -> count = 3
    - loop breaks

> In this program, all three blocks  if, elif, and else  are executed, but in different iterations of the loop, not all at once.

### ************ *Continue Statement in Python* ************

*Defination:*  
-> In Python, the continue statement is used inside loops to skip the current iteration and move to the next iteration immediately.
* The loop does not stop, only the code after continue in the current iteration is skipped.

  *Example:*

```python
for i in range(1, 6):
    if i == 3:
        continue  # skip the current iteration
    print(i)
```

> It is often used with if conditions to skip certain values.
